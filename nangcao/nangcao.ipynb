```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer 
data = load_breast_cancer()
data_df = pd.DataFrame(data = data.data, 
                       columns = data.feature_names) 
data_df.head().T
```

```python
from sklearn.datasets import load_breast_cancer 
cancer_dataset=load_breast_cancer()
cancer_dataset.keys()
```

```python
print(cancer_dataset.DESCR)
print(cancer_dataset.target_names)
# Attributes : Thuộc tính
# Instances : truong hop
```

```python
df=pd.DataFrame(cancer_dataset['data'],columns=cancer_dataset['feature_names'])
df.head()
```

PCA VISUALIZATION

```python
## Standardization
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()

scaler.fit(df)
scaled_data=scaler.transform(df)

scaled_data
```

```python
##  Applying PCA Algorithms
from sklearn.decomposition import PCA

pca=PCA(n_components=2)
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
print(scaled_data.shape)
print(x_pca.shape)
# tu 30 tinh nang xuong chi dung 2 tinh nang thoi nhung van con 569 du lieu tren hang
```

```python
plt.Figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=cancer_dataset['target'],cmap='plasma')
plt.xlabel("component dau tien")
plt.ylabel("component thu 2")
```
